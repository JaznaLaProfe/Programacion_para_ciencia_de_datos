#Lista

`````python
notas = [5.5, 6.0, 4.8, 5.9, 6.3]

print(notas)
print(f"Primera nota : {notas[0]}")
print(f"Promedio: {sum(notas)/len(notas)}")
`````

# Diccionario

`````python
estudiante = {
    "nombre": "Ana",
    "edad": 20,
    "carrera": "Ingeniería",
    "nota_final": 5.8
}

print(estudiante["nombre"])
print(estudiante["nota_final"])
`````

# Lista de diccionarios

`````python
estudiantes = [
    {"nombre": "Ana", "edad": 20, "nota": 5.8},
    {"nombre": "Luis", "edad": 22, "nota": 6.1},
    {"nombre": "María", "edad": 21, "nota": 5.4},
    {"nombre": "Pedro", "edad": 23, "nota": 4.9}
]

print(estudiantes[0])
print(estudiantes[2]["nombre"])
`````

**Problema que aparece**

Un problema típico en ciencia de datos:

+ Calcular promedios
+ Filtrar
+ Analizar columnas

Esto se vuelve incómodo.

# Array de `numpy`

`````python
notas = np.array([5.8, 6.1, 5.4, 4.9])

print(notas)
print(f"Promedio: {np.mean(notas)}")
print(f"Nota máxima: {np.max(notas)}")
`````

+ Operaciones matemáticas muy rápidas
+ Optimizado para datos numéricos

`````python
# Es posible hacer operaciones con vectores
notas_curva = notas + 0.2
print(notas_curva)
`````

**Problema que aparece**

*Y si se quiere guardar nombre, edad, carrera, nota. ¿Dónde se hace?*

# Dataframe

`````python
datos = {
    "nombre": ["Ana", "Luis", "María", "Pedro"],
    "edad": [20, 22, 21, 23],
    "nota": [5.8, 6.1, 5.4, 4.9]
}

data = pd.DataFrame(datos)

data
`````

## Operaciones típicas

### Promedio

`````python
data["nota"].mean()
`````

### Filtro de estudiantes aprobados

`````python
data[data["nota"] >= 4.0]
`````

### Agregar nuevas columnas

`````python
data["aprobado"] = data["nota"] >= 4.0
`````

### Algo un poco más elaborado ...

`````python
data.loc[data["nota"].idxmax()]
`````

# Ejercicio

Contexto

Una plataforma quiere recomendar películas a sus usuarios según el género que prefieren.

`````python
peliculas = [
    {"titulo": "Inception", "genero": "ciencia ficcion", "rating": 8.8},
    {"titulo": "Titanic", "genero": "romance", "rating": 7.9},
    {"titulo": "The Dark Knight", "genero": "accion", "rating": 9.0},
    {"titulo": "Interstellar", "genero": "ciencia ficcion", "rating": 8.6},
    {"titulo": "La La Land", "genero": "musical", "rating": 8.0},
    {"titulo": "Mad Max", "genero": "accion", "rating": 8.1}
]
`````

`````python
# Pregunta la preferencia del usuario
genero_preferido = input("¿Qué género de películas te gusta? ")
`````

## Buscar las recomendaciones

`````python
for p in peliculas:
    if p["genero"] == genero_preferido:
        print(f"Te recomendamos: {p['titulo']}")
`````

## Filtro por ranking

`````python
for p in peliculas:
  if p["genero"] == genero_preferido and p["rating"] > 8:
    print(f"Te recomendamos: {p['titulo']}, | rating:  {p['rating']}")
`````

## Crear una lista de recomendaciones

`````python
recomendadas = []

for p in peliculas:
    if p["genero"] == genero_preferido and p["rating"] > 8:
        recomendadas.append(p["titulo"])

print("Películas recomendadas:", recomendadas)
`````
